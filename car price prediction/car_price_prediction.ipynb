```python
print(newdata['fueltype'].unique())
print(newdata['doornumber'].unique())
```

```python
newdata['fueltype'] = newdata['fueltype'].map({'gas':0,'diesel':1})
newdata['doornumber']=newdata['doornumber'].map({'two':2,'four':4})
newdata['doornumber']
```

```python
from sklearn.preprocessing import LabelEncoder
```

### Step4: Converting Data into Numerical Data

```python
lbl=LabelEncoder()
```

```python
encodeddata = newdata.copy()
encodeddata[['carbody','drivewheel','fuelsystem']] = encodeddata[['carbody','drivewheel','fuelsystem']].apply(lambda x:lbl.fit_transform(x))
```

```python
encodeddata
```

```python
corr_matrix = encodeddata.corr()
plt.figure(figsize=(20,8))
sns.heatmap(corr_matrix,annot=True,cmap='plasma')
plt.title('Correlation Heatmap')
plt.show()
```

```python
y=encodeddata['price']
x=encodeddata.drop(columns={'price'})
```

### Step5: Training Data

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

### Step6: Linear Regression and Testing

```python
model.fit(xtrain,ytrain)
```

```python
predictions=model.predict(xtest)
```

```python
predictions
```

### Metrics For calculating the accuracy

```python
from sklearn.metrics import r2_score,mean_absolute_error, mean_squared_error
import math
```

```python

r2=r2_score(ytest,predictions)
r2=math.ceil(r2*100)
```

```python
print('The coefficient of determination is :--> ',r2,'%')
```

```python
mae = mean_absolute_error(ytest,predictions)
print('The Mean absolute error is :-->',mae)
```

```python
mse = mean_squared_error(ytest,predictions)
print('The mean squared error is :-->',mse)
```

```python
actual_predicted=pd.DataFrame({'Actual Price':ytest,'Predicted Price':predictions})
actual_predicted.head()
```

### Step7: Calculating the difference between the actual and predicted price 

```python
actual_predicted['error'] = actual_predicted['Predicted Price'] - actual_predicted['Actual Price']
actual_predicted.head()
```
